```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("used_car_price_dataset_extended.csv")  # Adjust the path if needed

# Check structure
df.info()
df.head()
```

```python
# Create a 2x2 grid of subplots
fig, axs = plt.subplots(2, 2, figsize=(18, 12))
# Graph 1: Count of Cars by Fuel Type
fuel_counts = df['fuel_type'].value_counts()
axs[0, 0].bar(fuel_counts.index, fuel_counts.values, color='skyblue')
axs[0, 0].set_title('Count of Cars by Fuel Type')
axs[0, 0].set_xlabel('Fuel Type')
axs[0, 0].set_ylabel('Number of Cars')

# Graph 2: Average Price by Transmission Type
avg_price_transmission = df.groupby('transmission')['price_usd'].mean()
axs[0, 1].bar(avg_price_transmission.index, avg_price_transmission.values, color='lightgreen')
axs[0, 1].set_title('Average Car Price by Transmission Type')
axs[0, 1].set_xlabel('Transmission')
axs[0, 1].set_ylabel('Average Price (USD)')

# Graph 3: Number of Cars per Brand (Top 10) - Horizontal Bar Chart
top_brands = df['brand'].value_counts().nlargest(10)
axs[1, 0].barh(top_brands.index, top_brands.values, color='salmon')
axs[1, 0].set_title('Top 10 Car Brands by Count')
axs[1, 0].set_xlabel('Number of Cars')
axs[1, 0].set_ylabel('Brand')


# Graph 4: Line Graph of Average Price Over Make Year
avg_price_per_year = df.groupby('make_year')['price_usd'].mean()
axs[1, 1].plot(avg_price_per_year.index, avg_price_per_year.values, marker='o', linestyle='-', color='purple')
axs[1, 1].set_title('Average Car Price Over the Years')
axs[1, 1].set_xlabel('Make Year')
axs[1, 1].set_ylabel('Average Price (USD)')
axs[1, 1].grid(True)

# Adjust layout
plt.tight_layout()
plt.show()
```
